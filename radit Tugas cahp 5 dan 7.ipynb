```python
import pandas as pd   # untuk membaca data dari file excel
from scipy.stats.mstats import gmean, hmean # untuk menghitung geometric mean dan harmonic mean 
import matplotlib.pyplot as plt # untuk membuat histogram dan menampilkan garis vertikal untuk masing masing mean

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #membaca data dari file excel dan menyimpan dalam variabel my_dataset

a_mean = my_dataset.Sr.mean() # menghitung arithmetic mean dari kolom Sr dalam dataset
g_mean = gmean(my_dataset['Sr']) # menghitung geometric mean dari kolom Sr dalam dataset
h_mean = hmean(my_dataset['Sr']) # menghitung harmonic mean dari kolom Sr dalam dataset

print ('-------') # mencetak garis pemisah 
print ('arithmetic mean') # mencetak judul untuk arithmetic mean 
print ("{0:.1f} [ppm]".format(a_mean)) # mencetak nilai arithmatic mean dengan format satu angka di belakang koma dan satuan ppm
print ('-------') # mencetak garis pemisah

print ('geometric mean') # mencetak judul untuk geometric mean
print ("{0:.1f} [ppm]".format(g_mean)) # mencetak nilai geometric mean dengan format satu angka di belakang koma dan satuan ppm
print ('-------') # mencetak garis pemisah

print ('harmonic mean') # mencetak judul untuk harmonic mean
print ("{0:.1f} [ppm]".format(h_mean)) # mencetak nilai harmonic mean dengan format satu angka di belakang koma dan satuan ppm
print ('-------') # mencetak garis pemisah

fig, ax = plt.subplots() # membuat figure dan axes untuk plot
ax.hist(my_dataset.Sr, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8) # membuat histogram dari kolom Sr dalam dataset dengan jumlah bin otomatis, menormalkan histogram, memberikan warna tepi pada setiap bin, memberikan label untuk legenda, dan mengatur transparansi histogram
ax.axvline(a_mean, color='purple', label='Arithmetic mean', linewidth=2) # menambahkan garis vertikal pada posisi arithmetic mean dengan warna ungu, label untuk legenda, dan ketebalan garis 2
ax.axvline(g_mean, color='pink',  label='Geometric mean', linewidth=2) # menambahkan garis vertikal pada posisi geometric mean dengan warna pink, label untuk legenda, dan ketebalan garis 2
ax.axvline(h_mean, color='green',  label='Harmonic mean', linewidth=2) # menambahkan garis vertikal pada posisi harmonic mean dengan warna hijau, label untuk legenda, dan ketebalan garis 2
ax.set_xlabel('Sr [ppm]') # mengatur label untuk sumbu x
ax.set_ylabel('Probability density') # mengatur label untuk sumbu y
ax.legend() # menampilkan legenda untuk garis vertikal yang menunjukkan masing-masing mean

''' 
Output: 
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import numpy as np # untuk menghitung histogram dan bin edges, serta untuk melakukan operasi matematika seperti penjumlahan dan perkalian pada array

def integrate_rec(f, a, b, n): # mendefinisikan fungsi integrate_rec yang menerima empat argumen: f (fungsi yang akan diintegrasikan), a (batas bawah integrasi), b (batas atas integrasi), dan n (jumlah subinterval untuk metode rectangle)
    # Implementation of the rectanlge method 
    h = (b-a)/n # menghitung lebar setiap subinterval dengan membagi selisih antara batas atas dan batas bawah dengan jumlah subinterval n
    x = np.linspace(a, b, n+1)  # membuat array x yang berisi titik-titik pembagi interval dari a hingga b dengan jumlah n+1 titik (termasuk batas bawah dan batas atas)
    i=0 # inisialisasi variabel i untuk digunakan sebagai indeks dalam perulangan
    area=0 # inisialisasi variabel area untuk menyimpan hasil integrasi yang akan dihitung dengan metode rectangle
    while i<n: # melakukan perulangan sebanyak n kali untuk menghitung luas dengan metode rectangle
        sup_rect =  f(x[i])*h # menghitung luas dari rectangle dengan menggunakan nilai fungsi f pada titik x[i] sebagai tinggi dan h sebagai lebar
        area += sup_rect # menambahkan luas rectangle yang dihitung pada iterasi saat ini ke variabel area yang menyimpan total luas yang dihitung sejauh ini
        i += 2  # increment nilai i untuk melanjutkan ke titik berikutnya pada array x
    return area # mengembalikan nilai area yang merupakan hasil integrasi yang dihitung dengan metode rectangle
'''
We test the Rectangle method on the sine funcion were the definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_rec(np.sin, 0, np.pi/2, 5) # menghitung hasil integrasi dari fungsi sinus pada interval [0, π/2] dengan menggunakan metode rectangle dan jumlah subinterval n=5
sup_10 = integrate_rec(np.sin, 0, np.pi/2, 10) # menghitung hasil integrasi dari fungsi sinus pada interval [0, π/2] dengan menggunakan metode rectangle dan jumlah subinterval n=10
sup_100 = integrate_rec(np.sin, 0, np.pi/2, 100) # menghitung hasil integrasi dari fungsi sinus pada interval [0, π/2] dengan menggunakan metode rectangle dan jumlah subinterval n=100

print('Using n=5, the rectangle method returns a value of {:.2f}'.format(sup_5)) # mencetak hasil integrasi yang dihitung dengan metode rectangle untuk n=5 dengan format dua angka di belakang koma
print('Using n=10, the rectangle method returns a value of {:.2f}'.format(sup_10)) # mencetak hasil integrasi yang dihitung dengan metode rectangle untuk n=10 dengan format dua angka di belakang koma
print('Using n=100, the rectangle method returns a value of {:.2f}'.format(sup_100)) # mencetak hasil integrasi yang dihitung dengan metode rectangle untuk n=100 dengan format dua angka di belakang koma
     
'''
Output:
Using n=5, the rectangle method returns a value of 0.83
Using n=10, the rectangle method returns a value of 0.92
Using n=100, the rectangle method returns a value of 0.99
'''
 
```

```python
import numpy as np # untuk menghitung histogram dan bin edges, serta untuk melakukan operasi matematika seperti penjumlahan dan perkalian pada array

def integrate_trap(f, a, b, n): # mendefinisikan fungsi integrate_trap yang menerima empat argumen: f (fungsi yang akan diintegrasikan), a (batas bawah integrasi), b (batas atas integrasi), dan n (jumlah subinterval untuk metode trapezoidal)
    # Implementation of the trapezoidal rule 
    h = (b-a)/n # menghitung lebar setiap subinterval dengan membagi selisih antara batas atas dan batas bawah dengan jumlah subinterval n
    x = np.linspace(a, b, n+1)  # membuat array x yang berisi titik-titik pembagi interval dari a hingga b dengan jumlah n+1 titik (termasuk batas bawah dan batas atas)
    i=1 # inisialisasi variabel i untuk digunakan sebagai indeks dalam perulangan
    area = h*(f(x[0]) + f(x[n]))/2 # menghitung luas awal dengan menggunakan nilai fungsi f pada titik x[0] dan x[n] sebagai tinggi dan h sebagai lebar, kemudian membagi hasilnya dengan 2 untuk mendapatkan rata-rata dari kedua nilai fungsi tersebut
    while i<n: # melakukan perulangan sebanyak n-1 kali untuk menghitung luas dengan metode trapezoidal
        sup_rect = f(x[i])*h # menghitung luas dari rectangle dengan menggunakan nilai fungsi f pada titik x[i] sebagai tinggi dan h sebagai lebar
        area += sup_rect  # menambahkan luas rectangle yang dihitung pada iterasi saat ini ke variabel area yang menyimpan total luas yang dihitung sejauh ini
        i += 1  # increment nilai i untuk melanjutkan ke titik berikutnya pada array x
    return area # mengembalikan nilai area yang merupakan hasil integrasi yang dihitung dengan metode trapezoidal

'''
We test the trapezoidal rule on the known sine funcion were the  
definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_trap(np.sin, 0, np.pi/2, 15) # menghitung hasil integrasi dari fungsi sinus pada interval [0, π/2] dengan menggunakan metode trapezoidal dan jumlah subinterval n=15
sup_10 = integrate_trap(np.sin, 0, np.pi/2, 20) # menghitung hasil integrasi dari fungsi sinus pada interval [0, π/2] dengan menggunakan metode trapezoidal dan jumlah subinterval n=20

print('Using n=5, the trapezoidal rule returns a value of {:.2f}'.format(sup_5)) # mencetak hasil integrasi yang dihitung dengan metode trapezoidal untuk n=5 dengan format dua angka di belakang koma
print('Using n=10, the trapezoidal rule returns a value of {:.2f}'.format(sup_10)) # mencetak hasil integrasi yang dihitung dengan metode trapezoidal untuk n=10 dengan format dua angka di belakang koma
   
'''
Output:
Using n=5, the trapezoidal rule returns a value of 0.99
Using n=10, the trapezoidal rule returns a value of 1.00
'''
 
```

```python
def simple_lithopress(z,ro=2900, g=9.8): # mendefinisikan fungsi simple_lithopress yang menerima tiga argumen: z (kedalaman dalam meter), ro (densitas batuan dalam kg/m^3, dengan nilai default 2900), dan g (percepatan gravitasi dalam m/s^2, dengan nilai default 9.8)
    pressure = z*g*ro/1e6 # menghitung tekanan hidrostatik dengan menggunakan rumus tekanan hidrostatik, dimana z adalah kedalaman, g adalah percepatan gravitasi, ro adalah densitas batuan, dan hasilnya dibagi dengan 1e6 untuk mengkonversi dari Pascal ke Megapascal (MPa)
    return pressure # mengembalikan nilai tekanan yang dihitung dengan rumus tekanan hidrostatik, dimana z adalah kedalaman, g adalah percepatan gravitasi, ro adalah densitas batuan, dan hasilnya dibagi dengan 1e6 untuk mengkonversi dari Pascal ke Megapascal (MPa)

my_pressure = simple_lithopress(z=3000) # menghitung tekanan pada kedalaman 3000 meter dengan menggunakan fungsi simple_lithopress yang telah didefinisikan, dimana z diatur ke 3000 meter dan nilai default untuk densitas batuan dan percepatan gravitasi digunakan
print('The pressure at 3000 meters is {0:.0f} MPa'.format(my_pressure)) # mencetak hasil perhitungan tekanan pada kedalaman 3000 meter dengan format tanpa angka di belakang koma dan satuan MPa

'''
Output: The pressure at 3000 meters is 87 MPa
'''
```

```python
import numpy as np # untuk menghitung histogram dan bin edges, serta untuk melakukan operasi matematika seperti penjumlahan dan perkalian pada array
from scipy.integrate import trapezoid # untuk menggunakan fungsi trapezoid dari scipy.integrate yang menyediakan metode untuk menghitung integral dengan menggunakan metode trapezoidal, dimana fungsi ini dapat digunakan untuk menghitung integral dari array yang berisi nilai-nilai fungsi pada titik-titik tertentu
import matplotlib.pyplot as plt # untuk membuat plot dari hasil perhitungan tekanan pada kedalaman tertentu dengan menggunakan fungsi simple_lithopress dan metode trapezoidal untuk menghitung integral dari densitas batuan terhadap kedalaman

r = np.linspace(1,6400,6400) # membuat array r yang berisi nilai-nilai dari 1 hingga 6400 dengan jumlah total 6400 elemen, dimana setiap elemen dalam array mewakili jarak dari pusat bumi dalam kilometer

def density(): # mendefinisikan fungsi density yang akan menghitung densitas batuan pada berbagai kedalaman di dalam bumi dengan menggunakan metode linspace untuk membuat array yang berisi nilai-nilai densitas pada setiap lapisan bumi, kemudian menggabungkan semua array tersebut menjadi satu array final yang berisi nilai-nilai densitas untuk seluruh kedalaman dari pusat bumi hingga permukaan
    ro_inner_core = np.linspace(14000, 12800, 1220) # membuat array ro_inner_core yang berisi nilai-nilai densitas untuk lapisan inti dalam bumi dengan menggunakan metode linspace untuk menghasilkan 1220 nilai yang terdistribusi secara merata antara 13100 kg/m^3 dan 12800 kg/m^3
    ro_outer_core = np.linspace(14700, 9900, 2259) # membuat array ro_outer_core yang berisi nilai-nilai densitas untuk lapisan inti luar bumi dengan menggunakan metode linspace untuk menghasilkan 2259 nilai yang terdistribusi secara merata antara 12200 kg/m^3 dan 9900 kg/m^3
    ro_lower_mantle = np.linspace(5600,4400,2171) # membuat array ro_lower_mantle yang berisi nilai-nilai densitas untuk lapisan mantel bawah bumi dengan menggunakan metode linspace untuk menghasilkan 2171 nilai yang terdistribusi secara merata antara 5600 kg/m^3 dan 4400 kg/m^3
    ro_upper_mantle = np.linspace(4130,3400,720) # membuat array ro_upper_mantle yang berisi nilai-nilai densitas untuk lapisan mantel atas bumi dengan menggunakan metode linspace untuk menghasilkan 720 nilai yang terdistribusi secara merata antara 4130 kg/m^3 dan 3400 kg/m^3
    ro_crust = np.linspace(3100,2700,30) # membuat array ro_crust yang berisi nilai-nilai densitas untuk lapisan kerak bumi dengan menggunakan metode linspace untuk menghasilkan 30 nilai yang terdistribusi secara merata antara 3100 kg/m^3 dan 2700 kg/m^3
   
    ro_final = np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust)) # menggabungkan semua array yang berisi nilai-nilai densitas untuk setiap lapisan bumi (ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, dan ro_crust) menjadi satu array final yang berisi nilai-nilai densitas untuk seluruh kedalaman dari pusat bumi hingga permukaan dengan menggunakan fungsi concatenate dari numpy
    
    return ro_final # mengembalikan array final yang berisi nilai-nilai densitas untuk seluruh kedalaman dari pusat bumi hingga permukaan yang telah dihitung dengan menggabungkan semua array yang berisi nilai-nilai densitas untuk setiap lapisan bumi

ro = density() # memanggil fungsi density yang telah didefinisikan untuk menghitung array ro yang berisi nilai-nilai densitas untuk seluruh kedalaman dari pusat bumi hingga permukaan

fig, ax = plt.subplots() # membuat figure dan axes untuk plot
ax.plot(r,ro, label="Densities on the Earth's Interior") # membuat plot dengan menggunakan fungsi plot dari matplotlib untuk menampilkan hubungan antara jarak dari pusat bumi (r) dan densitas batuan (ro) dengan memberikan label untuk legenda yang menunjukkan bahwa plot tersebut menggambarkan densitas pada interior bumi
ax.set_ylabel(r"Density [Kg/m$^3$]")  # mengatur label untuk sumbu y dengan menggunakan format LaTeX untuk menampilkan satuan kg/m^3 dengan superscript untuk angka 3
ax.set_xlabel("Distance from the Earth center r [km]") # mengatur label untuk sumbu x dengan teks yang menunjukkan bahwa sumbu x mewakili jarak dari pusat bumi dalam kilometer
ax.legend() # menampilkan legenda untuk plot yang menunjukkan bahwa garis pada plot tersebut mewakili densitas pada interior bumi
ax.grid() # menambahkan grid pada plot untuk memudahkan pembacaan nilai pada sumbu x dan sumbu y
```

```python
def gravity(r): # mendefinisikan fungsi gravity yang menerima argumen r (jarak dari pusat bumi dalam kilometer) dan akan menghitung percepatan gravitasi pada berbagai jarak dari pusat bumi dengan menggunakan rumus gravitasi Newton yang melibatkan konstanta gravitasi, densitas batuan, dan jarak dari pusat bumi, serta menggunakan metode trapezoidal untuk menghitung integral dari densitas batuan terhadap jarak untuk mendapatkan nilai percepatan gravitasi pada setiap jarak yang diberikan
    
    g = np.zeros(len(r)) # membuat array g yang berisi nilai-nilai percepatan gravitasi yang akan dihitung untuk setiap jarak dalam array r, dengan menginisialisasi semua nilai dalam array g menjadi 0 menggunakan fungsi zeros dari numpy dengan panjang yang sama dengan panjang array r
    Gr = 6.67408e-11 # konstanta gravitasi universal dalam satuan m^3 kg^-1 s^-2
    r = r * 1000 # mengkonversi jarak dari pusat bumi dari kilometer ke meter dengan mengalikan setiap elemen dalam array r dengan 1000
    
    for i in range(1,len(r)): # melakukan perulangan untuk setiap indeks i mulai dari 1 hingga panjang array r, dimana setiap iterasi akan menghitung percepatan gravitasi pada jarak r[i] dengan menggunakan rumus gravitasi Newton dan metode trapezoidal untuk menghitung integral dari densitas batuan terhadap jarak
        
        r1 = r[0:i] # membuat array r1 yang berisi nilai-nilai jarak dari pusat bumi dari indeks 0 hingga indeks i-1, dimana setiap elemen dalam array r1 mewakili jarak dari pusat bumi untuk setiap titik yang akan digunakan dalam perhitungan integral
        ro1  = ro[0:i] # membuat array ro1 yang berisi nilai-nilai densitas batuan yang sesuai dengan jarak dalam array r1, dimana setiap elemen dalam array ro1 mewakili densitas batuan pada jarak yang sesuai dengan elemen dalam array r1
        r2 = r1[i-2] # mengambil nilai jarak pada indeks i-2 dari array r1 untuk digunakan sebagai batas atas dalam perhitungan integral, dimana r2 mewakili jarak dari pusat bumi pada titik yang sedang dihitung percepatan gravitasinya
    
        y = ro1*r1**2 # menghitung nilai y yang merupakan hasil perkalian antara densitas batuan (ro1) dan kuadrat dari jarak (r1) untuk setiap elemen dalam array r1, dimana y mewakili nilai fungsi yang akan diintegrasikan untuk menghitung percepatan gravitasi pada jarak r2
        y_int = trapezoid(y,r1) # menghitung hasil integral dari array y terhadap r1 dengan menggunakan metode trapezoidal yang disediakan oleh fungsi trapezoid dari scipy.integrate, dimana y adalah nilai fungsi yang akan diintegrasikan dan r1 adalah titik-titik pembagi interval yang digunakan untuk metode trapezoidal, sehingga y_int mewakili hasil integral yang digunakan dalam perhitungan percepatan gravitasi pada jarak r2
        
        g1 = ((4 * np.pi*Gr)/(r2**2)) * y_int # menghitung nilai percepatan gravitasi g1 pada jarak r2 dengan menggunakan rumus gravitasi Newton yang melibatkan konstanta gravitasi (Gr), hasil integral (y_int), dan jarak r2, dimana g1 mewakili nilai percepatan gravitasi yang dihitung untuk jarak r2
        g[i] = g1 # menyimpan nilai percepatan gravitasi yang dihitung (g1) ke dalam array g pada indeks i, sehingga setiap elemen dalam array g akan berisi nilai percepatan gravitasi yang sesuai dengan jarak yang diberikan dalam array r
           
    return g # mengembalikan array g yang berisi nilai-nilai percepatan gravitasi yang dihitung untuk setiap jarak dalam array r dengan menggunakan rumus gravitasi Newton dan metode trapezoidal untuk menghitung integral dari densitas batuan terhadap jarak

g = gravity(r)  # memanggil fungsi gravity yang telah didefinisikan untuk menghitung array g yang berisi nilai-nilai percepatan gravitasi pada berbagai jarak dari pusat bumi dengan menggunakan rumus gravitasi Newton dan metode trapezoidal untuk menghitung integral dari densitas batuan terhadap jarak, dimana r adalah array yang berisi nilai-nilai jarak dari pusat bumi dalam kilometer

fig, ax = plt.subplots() # membuat figure dan axes untuk plot
ax.plot(r,g) # membuat plot dengan menggunakan fungsi plot dari matplotlib untuk menampilkan hubungan antara jarak dari pusat bumi (r) dan percepatan gravitasi (g) dengan memberikan label untuk legenda yang menunjukkan bahwa garis pada plot tersebut mewakili percepatan gravitasi pada berbagai jarak dari pusat bumi
ax.grid() # menambahkan grid pada plot untuk memudahkan pembacaan nilai pada sumbu x dan sumbu y
ax.set_ylabel(r'g [m/s$^2]$') # mengatur label untuk sumbu y dengan menggunakan format LaTeX untuk menampilkan satuan m/s^2 dengan superscript untuk angka 2
ax.set_xlabel('Distance from the Earth center r [km]') # mengatur label untuk sumbu x dengan teks yang menunjukkan bahwa sumbu x mewakili jarak dari pusat bumi dalam kilometer
```

```python
def pressure(r, ro, g): # mendefinisikan fungsi pressure yang menerima tiga argumen: r (jarak dari pusat bumi dalam kilometer), ro (array yang berisi nilai-nilai densitas batuan pada berbagai kedalaman), dan g (array yang berisi nilai-nilai percepatan gravitasi pada berbagai jarak dari pusat bumi), dimana fungsi ini akan menghitung tekanan pada berbagai kedalaman di dalam bumi dengan menggunakan rumus tekanan hidrostatik yang melibatkan integral dari hasil perkalian antara densitas batuan dan percepatan gravitasi terhadap jarak dari pusat bumi, serta menggunakan metode trapezoidal untuk menghitung integral tersebut
    
    p = np.zeros(len(r))    # membuat array p yang berisi nilai-nilai tekanan yang akan dihitung untuk setiap jarak dalam array r, dengan menginisialisasi semua nilai dalam array p menjadi 0 menggunakan fungsi zeros dari numpy dengan panjang yang sama dengan panjang array r
    r = r *1000 # mengkonversi jarak dari pusat bumi dari kilometer ke meter dengan mengalikan setiap elemen dalam array r dengan 1000
    
    for i in range(0,len(r)): # melakukan perulangan untuk setiap indeks i mulai dari 0 hingga panjang array r, dimana setiap iterasi akan menghitung tekanan pada jarak r[i] dengan menggunakan rumus tekanan hidrostatik yang melibatkan integral dari hasil perkalian antara densitas batuan dan percepatan gravitasi terhadap jarak dari pusat bumi, serta menggunakan metode trapezoidal untuk menghitung integral tersebut    
        r1 = r[i:len(r)] # membuat array r1 yang berisi nilai-nilai jarak dari pusat bumi mulai dari indeks i hingga indeks terakhir, dimana setiap elemen dalam array r1 mewakili jarak dari pusat bumi untuk setiap titik yang akan digunakan dalam perhitungan integral
        ro1 = ro[i:len(r)] # membuat array ro1 yang berisi nilai-nilai densitas batuan yang sesuai dengan jarak dalam array r1, dimana setiap elemen dalam array ro1 mewakili densitas batuan pada jarak yang sesuai dengan elemen dalam array r1
        g1 = g[i:len(r)] # membuat array g1 yang berisi nilai-nilai percepatan gravitasi yang sesuai dengan jarak dalam array r1, dimana setiap elemen dalam array g1 mewakili percepatan gravitasi pada jarak yang sesuai dengan elemen dalam array r1
        y = ro1*g1 # menghitung nilai y yang merupakan hasil perkalian antara densitas batuan (ro1) dan percepatan gravitasi (g1) untuk setiap elemen dalam array r1, dimana y mewakili nilai fungsi yang akan diintegrasikan untuk menghitung tekanan pada jarak r[i]
        p1 = trapezoid(y,r1) # menghitung hasil integral dari array y terhadap r1 dengan menggunakan metode trapezoidal yang disediakan oleh fungsi trapezoid dari scipy.integrate, dimana y adalah nilai fungsi yang akan diintegrasikan dan r1 adalah titik-titik pembagi interval yang digunakan untuk metode trapezoidal, sehingga p1 mewakili hasil integral yang digunakan dalam perhitungan tekanan pada jarak r[i]
        p[i] = p1 # menyimpan nilai tekanan yang dihitung (p1) ke dalam array p pada indeks i, sehingga setiap elemen dalam array p akan berisi nilai tekanan yang sesuai dengan jarak yang diberikan dalam array r
    return p # mengembalikan array p yang berisi
    
p = pressure(r,ro,g)/1e9 # memanggil fungsi pressure yang telah didefinisikan untuk menghitung array p yang berisi nilai-nilai tekanan pada berbagai kedalaman di dalam bumi dengan menggunakan rumus tekanan hidrostatik yang melibatkan integral dari hasil perkalian antara densitas batuan dan percepatan gravitasi terhadap jarak dari pusat bumi, serta menggunakan metode trapezoidal untuk menghitung integral tersebut, dimana r adalah array yang berisi nilai-nilai jarak dari pusat bumi dalam kilometer, ro adalah array yang berisi nilai-nilai densitas batuan pada berbagai kedalaman, dan g adalah array yang berisi nilai-nilai percepatan gravitasi pada berbagai jarak dari pusat bumi, kemudian hasilnya dibagi dengan 1e9 untuk mengkonversi dari Pascal ke GigaPascal (GPa)
z = np.linspace(6400, 1, 6400) # membuat array z yang berisi nilai-nilai kedalaman dari permukaan bumi ke pusat bumi dengan menggunakan metode linspace untuk menghasilkan 6400 nilai yang terdistribusi secara merata antara 6400 km (kedalaman maksimum) dan 1 km (kedalaman minimum), dimana setiap elemen dalam array z mewakili kedalaman dari permukaan bumi dalam kilometer

fig, ax = plt.subplots() # membuat figure dan axes untuk plot
ax.plot(z,p) # membuat plot dengan menggunakan fungsi plot dari matplotlib untuk menampilkan hubungan antara kedalaman dari permukaan bumi (z) dan tekanan (p) dengan memberikan label untuk legenda yang menunjukkan bahwa garis pada plot tersebut mewakili tekanan pada berbagai kedalaman di dalam bumi
ax.grid() # menambahkan grid pada plot untuk memudahkan pembacaan nilai pada sumbu x dan sumbu y
ax.set_ylabel('P [GPa]') # mengatur label untuk sumbu y dengan teks yang menunjukkan bahwa sumbu y mewakili tekanan dalam satuan GigaPascal (GPa)
ax.set_xlabel('Depth z from the Earth Surface [km]') # mengatur label untuk sumbu x dengan teks yang menunjukkan bahwa sumbu x mewakili kedalaman dari permukaan bumi dalam kilometer
```
