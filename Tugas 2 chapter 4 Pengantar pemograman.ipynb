```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def my_line(x, m, q):
    y = m * x + q
    return y

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_majors', engine='openpyxl')

my_dataset2 = my_dataset[my_dataset.Epoch == 'two']

x_line = np.linspace(52.5, 62, 100)
y_line = my_line(x_line, m=0.3, q=-10.3)

fig, ax = plt.subplots(figsize=(10, 7))

ax.scatter(my_dataset2.SiO2, my_dataset2.CaO, marker='o', color="#2ce23b", edgecolor='#000000', label=r'$2^{nd}$ Epoch')

ax.plot(x_line, y_line, color='#342a77', linewidth=2)


ax.annotate(r'What is the $1\sigma$ for this point?', 
            xy=(47.6, 6.6), xytext=(45, 9.0), 
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3"),
            fontsize=10)

line_label = r'$Na_2O = 0.3 \cdot SiO_2 - 10.3$'
ax.text(52.4, 5.6, line_label, dict(size=10, rotation=33))

mu_val = 57.721
mu_formula = r'$\mu_{SiO_2} = \frac{a_1+a_2+\dots+a_n}{n} = ' + f'{mu_val:.1f}' + r'\text{ [wt.%]}$'
ax.text(51.5, 5.1, mu_formula, dict(size=11))

ax.set_xlabel(r'$SiO_2$ [wt.%]')
ax.set_ylabel(r'$K_2O$ [wt.%]') 

ax.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
